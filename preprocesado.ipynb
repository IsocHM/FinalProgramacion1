```python
!pip install ucimlrepo

import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo

mushroom = fetch_ucirepo(id=73)

X = mushroom.data.features
y = mushroom.data.targets

df = pd.DataFrame(X)

#Completar datos faltantes
np.random.seed(0)
missing_rate = 0.05
for column in df.columns:
    n_missing = int(np.floor(missing_rate * df.shape[0]))
    missing_indices = np.random.choice(df.index, n_missing, replace=False)
    df.loc[missing_indices, column] = np.nan
    #Imputar datos con la moda
    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)

#Verificar datos faltantes
print("\nDatos faltantes después de la imputación:")
print(df.isnull().sum())

#Convertir variables categoricas
df_encoded = pd.get_dummies(df, drop_first=True)

X_processed = df_encoded
y_processed = pd.Series(y['poisonous'])

print("\nDatos después del preprocesado:")
print(X_processed.head())
print("\nVariable objetivo:")
print(y_processed.head())

X_processed.to_csv('mushroom_preprocessed.csv', index=False)
y_processed.to_csv('mushroom_targets.csv', index=False)

print("\nEl dataset ha sido preprocesado y guardado en datasets diferentes")
```
